```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json
```

```python
data = pd.read_csv('../../output/tests/text/stats.csv')

data.head()
```

```python
# Plot Frame Type vs Avg Cosine Similarity and Avg Base Cosine Similarity
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Avg Cosine Similarity', y='Avg Base Cosine Similarity', 
                hue='Frame Type', data=data, s=100)
plt.title('Avg Cosine Similarity vs Avg Base Cosine Similarity by Frame Type')
plt.xlabel('Average Cosine Similarity')
plt.ylabel('Average Base Cosine Similarity')
plt.legend(title='Frame Type')
# save
plt.savefig('cosine_similarity.png')
plt.show()
```
